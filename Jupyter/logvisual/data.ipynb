```python
logFolder = "./logs"

def logFile2DataFrame( filePath ) :
    dfLog = pd.DataFrame( columns=[ 'date' , 'timestamp' , 'processID' , 'level' , 'message' ] )
    datePattern = "[0-9]{2}\/[0-9]{2}\/[0-9]{4}"

    with open( filePath , 'r' , encoding='UTF8') as logFile :
        numRows = -1
        for line in logFile :
            if re.search( datePattern , line ) :
                tokens    = line.split(" ")
                date = tokens[0]
                timestamp = tokens[1]
                processID = tokens[2]
                level     = tokens[3]
                message   = ",".join( tokens[4:] )
                numRows += 1
                dfLog.loc[ numRows ] = [ date , timestamp , processID , level , message ]
            else :
                # Multiline message, integrate it into last record
                dfLog.loc[ numRows , message ] += line
    return dfLog
```
